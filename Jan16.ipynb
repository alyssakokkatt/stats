```python
scores = [65,78,30,85]

def get_grade(score):
  if score >= 71 and score <= 100:
    return "A"
  elif score >= 61 and score <= 70:
    return "C"
  elif score >= 50 and score <= 60:
    return "D"
  elif score >= 40 and score < 50:
    return "E"
  else:
    return "U"

for s in scores:
  print(f"s={s}, Grade: {get_grade(s)}")
```
